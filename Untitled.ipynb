```python
plt.scatter(df['Gross national income (GNI) per capita'], df['Life expectancy at birth'], color='c')
plt.xlabel('Gross national income (GNI) per capita')
plt.ylabel('Life expectancy at birth')
plt.show()
```

Just as I suspected. From 0 to 200000 GNI there is a very steep increase in life expectancy, rising from 55 to 75+ an increase in 20 years! After this life expectancy seems to plateau, suggesting that after a certain point GNI ceases to influence life expectancy. 

```python
def to_number(s):
    try:
        return float(s)
    except ValueError:
        return None

years = df.columns.tolist()[-8:]

trends = pd.DataFrame(columns=['Country', 'Year', 'HDI'])

    
for year in years:
    for index, row in df.iterrows():
        vals = {'Country': row['Country'], 'Year': int(year), 'HDI':to_number(row[year])}
        trends = trends.append(vals, ignore_index=True)    
```

```python
traces = []

#create a trace for each country 
for country in trends['Country'].unique():
    traces.append(go.Scatter(
        x=trends.loc[trends['Country']==country]['Year'].tolist(),
        y=trends.loc[trends['Country']==country]['HDI'].tolist(),
        mode='lines',
        connectgaps=False,
        name = country
    ))

layout = go.Layout(
    width=800,
    height=1000,
    xaxis=dict(
        showline=True,
        showgrid=False,
        showticklabels=True,
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        ticks='outside',
        tickcolor='rgb(204, 204, 204)',
        tickwidth=2,
        ticklen=5,
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(82, 82, 82)',
        ),
        title='Year'
    ),
    yaxis=dict(
        showgrid=True,
        zeroline=False,
        showline=False,
        showticklabels=True,
        title='HDI'
    ),
    autosize=False,
    margin=dict(
        autoexpand=False,
        l=100,
        r=20,
        t=110,
    ),
    showlegend=False
)


fig = go.Figure(data=traces, layout=layout)
iplot(fig, filename='HDI_trends')
```

As we can see from the graph above, most countries have been steadily increasing in Human Development Index at a relatively uniform rate. By exploring this dataset we can find instances where the HDI drops significantly. One such example is Syria, which was steadily increasing in HDI up until 2012, around the same time the <a href='https://www.britannica.com/event/Syrian-Civil-War'>Syrian civil war began escalating.</a> There are other examples in this dataset where war has coincided with declines in HDI, such is the case for Yemen which saw a <a href='https://en.wikipedia.org/wiki/Yemeni_Civil_War_(2015%E2%80%93present)'>civil war begin in 2015.</a> It is no surprise that mass conflict would have a negative impact on HDI, though there may be other factors which compund these issues such as international sanctions.
